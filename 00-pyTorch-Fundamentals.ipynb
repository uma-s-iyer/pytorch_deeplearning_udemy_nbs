## Reproducibility: (Trying to take the random out of random)

In short how a neural network learns i s
`astart wuith random numbers -. tensor operations -> update the random numbers to try and talke them to make better of the data ->l again -> again ->again `

To reduce the randomness in newural networks and pytorch comes the convept of a **random seed**

Essentially what the random seed does is "flavour" the randomness and gives a starting point gfor generating the randomness

**psuedo-randomness**
* https://pytorch.org/docs/stable/notes/randomness.html

```python
random_tensor_A = torch.rand(3,3)
random_tensor_B = torch.rand(3,3)
print(random_tensor_A)
print(random_tensor_B)

print(random_tensor_B == random_tensor_A)
#here you notive theya re never equeal and true random
```

```python
import torch
RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)

random_tensor_C = torch.rand(3,3)
#note: you need to provide the seed again before any random generation else it does not get applied to the random generator
torch.manual_seed(RANDOM_SEED)
random_tensor_D = torch.rand(3,3)
print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)
```
